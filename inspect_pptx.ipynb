```python
import win32com.client
import os

PPTX_PATH = os.path.abspath(r"C:\Latihan\Training-SQL-Server\contents\Module-00-Course-Introduction.pptx")
print("Opening:", PPTX_PATH)

app = win32com.client.Dispatch("PowerPoint.Application")
app.Visible = True
prs = app.Presentations.Open(PPTX_PATH, ReadOnly=True, Untitled=False, WithWindow=True)

print(f"Slide count: {prs.Slides.Count}")
print(f"Width: {prs.PageSetup.SlideWidth:.2f} pts  Height: {prs.PageSetup.SlideHeight:.2f} pts")
print()
```

```python
from pprint import pprint

PT = 72  # points per inch

for si in range(1, prs.Slides.Count + 1):
    slide = prs.Slides(si)
    print(f"{'='*70}")
    print(f"SLIDE {si}")
    print(f"{'='*70}")
    for shi in range(1, slide.Shapes.Count + 1):
        sh = slide.Shapes(shi)
        l = sh.Left / PT
        t = sh.Top / PT
        w = sh.Width / PT
        h = sh.Height / PT
        txt = ""
        if sh.HasTextFrame:
            try:
                txt = sh.TextFrame.TextRange.Text.replace("\r", " ").replace("\n", " ")
                if len(txt) > 80:
                    txt = txt[:80] + "..."
            except:
                txt = "<err>"
        visible = sh.Visible
        print(f"  #{shi:02d} L={l:.2f}\" T={t:.2f}\" W={w:.2f}\" H={h:.2f}\"  vis={visible}  '{txt}'")
    print()
```

```python
# Focus on timetable slides (slides 4 and 5) — check column alignment
# Expected column starts (inches): TIME~0.6, MODULE~2.15, FOCUS~5.0, DURATION~10.55
SLIDE_INDICES = [4, 5]

for si in SLIDE_INDICES:
    slide = prs.Slides(si)
    print(f"\n{'='*70}")
    print(f"TIMETABLE SLIDE {si} — shapes sorted by Top then Left")
    print(f"{'='*70}")
    shapes = []
    for shi in range(1, slide.Shapes.Count + 1):
        sh = slide.Shapes(shi)
        if sh.HasTextFrame:
            try:
                t_val = sh.TextFrame.TextRange.Text.strip()
            except:
                t_val = ""
            shapes.append((sh.Top/PT, sh.Left/PT, sh.Width/PT, sh.Height/PT, t_val))
    shapes.sort()
    for top, left, w, h, txt in shapes:
        short = (txt[:60] + "...") if len(txt) > 60 else txt
        print(f"  T={top:.2f}\" L={left:.2f}\" W={w:.2f}\" H={h:.2f}\"  |{short}|")
```

```python
prs.Close()
app.Quit()
print("Done — PowerPoint closed.")
```
